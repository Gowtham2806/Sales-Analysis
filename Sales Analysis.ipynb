### Merging 12 months of sales data into a single file

### Clean up the data

```python
nan_df = df[df.isna().any(axis=1)]
```

```python
nan_df
```

```python
df = df.dropna(axis=0, how='all')
```

```python
nan_df = df[df.isna().any(axis=1)]
```

```python
nan_df
```

```python
df = df[df['Order Date'].str[0:2] != 'Or']
df.head()
```

```python
temp_df = df[df['Order Date'].str[0:2] == 'Or']
temp_df.head()
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
df['Price Each'] = pd.to_numeric(df['Price Each'])

print(df['Price Each'].dtype)
print(df['Quantity Ordered'].dtype)
```

### Adding additional columns

#### Adding Months Column

```python
df['Months'] = df['Order Date'].str[0:2]
df['Months'] = df['Months'].astype('int32')

df.head()
```

```python
df
```

#### Adding Sales Column

```python
df['Sales'] = df['Quantity Ordered'] * df['Price Each']
```

```python
df.head(10)
```

#### Adding City Column

```python
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]


df['City'] = df['Purchase Address'].apply(lambda x: get_city(x)+ ' ' +get_state(x))

df.head()
```

### Question 4: What products are most often sold together?

```python
new_df = df[df['Order ID'].duplicated(keep=False)]

new_df.head(20)
```

```python
new_df
```

```python
new_df['Grouped'] = new_df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

new_df.head()

new_df = new_df[['Order ID', 'Grouped']].drop_duplicates()

new_df
```

```python
df.drop(columns='Grouped', inplace=True)
```

```python
df
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in new_df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    

for key, value in count.most_common(10):
    print(key, value)
```
